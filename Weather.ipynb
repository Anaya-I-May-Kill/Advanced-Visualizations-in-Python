```python
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
```

```python
weather = pd.read_csv('Weather Dataset.csv')
weather.head(5)
```

```python
weather.info()
```

```python
sns.barplot(x = weather['wind_speed'] , y = weather['temperature'])
```

```python
sns.distplot(weather['humidity'] , rug = False)
```

```python
sns.jointplot(x = weather['humidity'] , y = weather['temperature'] , kind = 'hex')
```

```python
sns.jointplot(x = weather['humidity'] , y = weather['temperature'] , kind = 'kde')
```

```python
sns.pairplot(weather[['humidity' , 'temperature' , 'air_pollution_index']])
```

```python
sns.set_palette("Paired")
sns.stripplot(x = weather['weather_type'] , y = weather['temperature'])
```

```python
sns.swarmplot(x = weather['humidity'] , y = weather['temperature'])
```

```python
sns.boxplot(x = weather['humidity'] , y = weather['temperature'] , hue = weather['weather_type'])
```

```python
sns.barplot(x = weather['humidity'] , y = weather['temperature'] , hue = weather['weather_type'])
```

```python
sns.pointplot(x = weather['humidity'] , y = weather['temperature'] , hue = weather['weather_type'])
```

```python
sns.lmplot(x = 'humidity' , y = 'temperature' , data = weather)
```

```python
sns.lmplot(x = 'humidity' , y = 'temperature' , hue = 'weather_type' , data = weather)
```
